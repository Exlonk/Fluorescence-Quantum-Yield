<h4 style="color:orange;"> Charge </h4>

Its necessary to add the charge of every molecule to the dataframe

```python
# Example SMILES string
# example = 'O=C([O-])c1ccccc1-c1c2ccc(=O)cc-2oc2cc([O-])ccc12' -> -2  # Choline, which has a total charge of +1
charges = []
for i in plqy_unique['SMILES']:
    # Convert the SMILES string to an RDKit molecule object
    mol = Chem.MolFromSmiles(i)

    # Calculate the total charge by summing the formal charge of each atom
    total_charge = sum(atom.GetFormalCharge() for atom in mol.GetAtoms())
    charges.append(total_charge)
```

```python
plqy_unique['CHARGE'] = charges
```

```python
# Filter the molecules with a charge greater than 0
plqy_unique[plqy_unique['CHARGE']>0]
```

Copy the xyz files that has boro and fluor to in a new folder

```python
def linux_command_queue(command):
    process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    # Wait for the process to complete and capture output/errors
    # Capture the output and error, if any
    output, error = process.communicate()

    # Decode the output and error from bytes to strings
    output_str = output.decode('utf-8').strip()
    error_str = error.decode('utf-8').strip()

    return output_str,error_str
```

```python
id_subset = pd.read_csv('/home/exl/Documents/Projects/Doctorade/Thesis/Results/database_bf/id_database_bf.csv')
id_subset
```

```python
origin = '/home/exl/Documents/Projects/Doctorade/Thesis/Results/database/structures/xyz/'
destination = '/home/exl/Documents/Projects/Doctorade/Thesis/Results/database_bf/structures/xyz'
command = "cp {0} {1}".format(origin,destination)

for i in id_subset['ID']:
    command = "cp {0} {1}".format(origin+str(i)+'.*',destination)
    linux_command_queue(command)

    command = "cp {0} {1}".format(origin+'s0_'+str(i)+'.*',destination)
    linux_command_queue(command)

    
```
